```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
K=10
rand_data = np.random.rand(2,100) #2차원 100개 data
x = np.random.rand(2,1) #2차원 1개 data
y = np.random.rand(2,1)
z = np.random.rand(2,1)
```

```python
dataLen = rand_data.shape[1]

if K < dataLen:
    pass
else:
    K = dataLen
```

### Euclidean distance ###

```python
#euclidean
euclidean_distance = np.sqrt(((rand_data-x)**2).sum(axis=0))#차원이 달라도 빼기 가능
euclidean_idx = np.argsort(euclidean_distance) #정렬하고 index 값 가져오기
euclidean_neighborIndex = euclidean_idx[:K] #K개 데이터 추출
```

```python
plt.plot(rand_data[0, :], rand_data[1, :], "ok", label = "Data") #검정색 점
plt.plot(x[0,0], x[1,0], "or", label = "Query") #기준점
plt.plot(rand_data[0, euclidean_neighborIndex], rand_data[1, euclidean_neighborIndex], "o", 
         markerfacecolor = "None", markersize = 12,
         markeredgewidth = 0.5, label = "KNN") #기준점 마킹
plt.legend(loc = 'upper right')
plt.title('KNN search(euclidean 1)')
plt.show()
```

### Cosine distance ###

```python
#cosine
from numpy import dot
from numpy.linalg import norm

_rand_data = np.transpose(rand_data)
cosine_distance = np.zeros(shape=(100,))
for i in range(len(_rand_data)):
    cosine_distance[i] = 1. - dot(_rand_data[i],x)/(norm(_rand_data[i])*norm(x))    
cosine_idx = np.argsort(cosine_distance)
cosine_neighborIndex = cosine_idx[:K]
```

```python
plt.plot(rand_data[0, :], rand_data[1, :], "ok", label = "Data") #검정색 점
plt.plot(x[0,0], x[1,0], "or", label = "Query") #기준점
plt.plot(rand_data[0, cosine_neighborIndex], rand_data[1, cosine_neighborIndex], "o", 
         markerfacecolor = "None", markersize = 12,
         markeredgewidth = 0.5, label = "KNN") #기준점 마킹
plt.legend(loc = 'upper right')
plt.title('KNN search(cosine1)')
plt.show()
```
